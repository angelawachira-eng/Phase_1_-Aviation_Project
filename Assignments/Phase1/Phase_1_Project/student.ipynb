### OBJECTIVE 1
Which combination of Make and Model offers the highest level of occupant safety and the lowest risk of total loss?

```python
#Analysis by Make only. Checking which make has most accidents.
make_counts = new_df['Make'].value_counts().head(10)
#Visual 1
plt.figure(figsize=(12, 6))
sb.barplot(x=make_counts.index, y=make_counts.values, palette='Blues_r')
plt.title('Total Accidents by  Make', fontsize=14)
plt.xlabel('Make')
plt.ylabel('Count of Accidents')
plt.xticks(rotation=45)
plt.show()
```

The analysis above shows the Total number of accidents by Make.
Makes like Cessna and Piper have very high accident counts because they produce thousands of small private aircraft that fly millions of hours annually.

```python
#Grouping Make,Model, Engine Type and Aircraft Category to discover aircraft with highest occupant safety & lowest risk of loss.
grouped = new_df.groupby(['Make', 'Model', 'Engine.Type', 'Aircraft.Category'])

# Aggregate statistics 

stats = grouped.agg(Total_Accidents=('Aircraft.damage', 'count'),Total_Occupants=('Total_Occupants', 'sum'),Total_Fatalities=('Total.Fatal.Injuries', 'sum'),Destroyed_Count=('Aircraft.damage', lambda x: (x == 'Destroyed').sum())).reset_index()
# A minimum of 100 accidents ensures statistical reliable safety comparisons and avoids misleading conclusions driven by small-sample anomalies.
min_accidents = 100

significant_stats = stats[(stats['Total_Accidents'] >= min_accidents) &(stats['Total_Occupants'] > 0)].copy()


# Safety Metrics
significant_stats['Fatality_Rate'] = (significant_stats['Total_Fatalities'] /significant_stats['Total_Occupants'])

significant_stats['Survival_Rate'] = 1 - significant_stats['Fatality_Rate']

significant_stats['Total_Loss_Risk'] = (significant_stats['Destroyed_Count'] /significant_stats['Total_Accidents'])

# Top 10 Safest Combinations
top_safety = significant_stats.sort_values(by=['Survival_Rate', 'Total_Loss_Risk'],ascending=[False, True]).head(10)

                                        
#Visual 2

plt.figure(figsize=(12, 8))

top_safety['Label'] = (top_safety['Make'] + ' ' + top_safety['Model'])

sb.barplot(data=top_safety,x='Survival_Rate',y='Label')

plt.title(f'Top 10 Safest Aircraft Combinations\n'f'(Min {min_accidents} Accidents)',fontsize=16)
plt.xlabel('Survival Rate')
plt.ylabel('Aircraft')
plt.xlim(0.90, 1.0)
plt.tight_layout()
plt.show()
```

The analysis above identifies aircraft combinations with the highest levels of occupant safety and the lowest risk of total loss by examining historical accident data grouped by Make, Model, Engine Type, and Aircraft Category. After applying a minimum accident threshold of 100 to ensure statistical reliability, survival rate and total loss risk emerged as effective indicators of overall safety performance.

### OBJECTIVE 2

### OBJECTIVE 3
Which aircraft categories provide the highest chance of survival in the event of an accident?

The code and visual shows how safe different aircraft categories are, by examining how often occupants survive accidents within each category. By grouping the data by aircraft category and averaging survival rates, it becomes easier to see which types of aircraft generally provide better protection for occupants when accidents occur.

```python
# Group by Category and calculate average survival rate
category_safety = new_df.groupby('Aircraft.Category')['Survival_Rate'].mean().sort_values(ascending=False).reset_index()

#Visual
plt.figure(figsize=(10, 5))
sb.barplot(data=category_safety, x='Survival_Rate', y='Aircraft.Category', palette='Blues_r')
plt.title('Objective 2: Survival Rate by Aircraft Category', fontsize=14)
plt.xlabel('Average Survival Rate (1.0 = 100% Survival)', fontsize=12)
plt.xlim(0, 1.1)
plt.show()
```

### OBJECTIVE 4
Which Engine types provide the most reliable performance?

```python
An analysis on how occupant safety varies across different engine types by looking at average survival rates.Highlights which types of aircraft engines have generally better occupant survival outcomes.The bar chart makes it easy to compare engine types to see which ones tend to perform better regarding safety.
```

```python
# Group by Engine Type
engine_safety = new_df.groupby('Engine.Type')['Survival_Rate'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 5))
sb.barplot(data=engine_safety, x='Survival_Rate', y='Engine.Type', palette='Oranges_r')
plt.title('Objective 3: Survival Rate by Engine Type', fontsize=14)
plt.xlabel('Average Survival Rate', fontsize=12)
plt.show()
```
